```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1 : Read the training data

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

### Step 2 Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X and Y (species)

```python
X = df.drop(columns=['species'])
Y =df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Step 4: Apply preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y,
                                                test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Build the decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=1,
                               min_samples_split=5,
                               min_samples_leaf=5,
                               criterion='gini')
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Step 7: Perform Hypeparameter tuning

```python
params = {'max_depth':[1, 2, 3, 4, 5],
          'min_samples_split':[3, 4, 5, 6, 7],
          'min_samples_leaf':[5, 6, 7, 8, 9],
          'criterion':['gini', 'entropy']}
```

```python
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier()
gscv = GridSearchCV(dtc, param_grid=params, cv=5, scoring='f1_macro')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

### Model Evaluation

```python
# confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```

```python
# Predict results for train and test
ypred_train = best_dtc.predict(xtrain)
ypred_test = best_dtc.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
# Print classification report 
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
# cross validate on train
from sklearn.model_selection import cross_val_score
scores_f1 = cross_val_score(best_dtc, xtrain, ytrain, cv=5, scoring='f1_macro')
scores_f1
```

```python
scores_f1.mean()
```

```python
scores_recall = cross_val_score(best_dtc, xtrain, ytrain, 
                                cv=5, scoring='recall_macro')
scores_recall
```

```python
scores_recall.mean()
```

```python
scores_precision = cross_val_score(best_dtc, xtrain, ytrain, 
                                cv=5, scoring='precision_macro')
scores_precision
```

```python
scores_precision.mean()
```

```python
scores_acc = cross_val_score(best_dtc, xtrain, ytrain,
                             cv=5, scoring='accuracy')
scores_acc
```

```python

scores_acc.mean()
```
