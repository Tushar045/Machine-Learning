### Step 1: Read the Dataset

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

### Step 3 : Seperate X and Y(species)
species ~ sep_len, sep_wid, pet_len, pet_wid


```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

### Step 4 : Apply Preprocessing on X

```python
X.dtypes
```

Because all features are numerical i will create a numerical pipeline only

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5 : Apply train test split

test size = 0.33

```python

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, 
                                                test_size=0.33, 
                                                random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6 : Building the Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python

model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

```python
c = X.corr()
c
```

```python
import seaborn as sns
sns.heatmap(data=c, annot=True, fmt='.2f')
```

### Step 7: Model evaluation

```python

# train accuracy
model.score(xtrain, ytrain)
```

```python
# test accuracy
model.score(xtest, ytest)
```

```python

# Confusion Matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

### Predict results for xtrain, xtest

```python

ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python

ytest.head()
```

```python
# Print the classification reports
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

### On test data i have macro f1_score of 0.94 which is very good model
### Cross validate results on train data

```python

from sklearn.model_selection import cross_val_score 
scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring='f1_macro')
scores
```

```python
scores.mean()
```

### Cross validated f1 macro score is 0.9703 which is a good score
