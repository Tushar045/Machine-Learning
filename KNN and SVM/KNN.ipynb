### KNN - (K-Nearest_Neigbors)

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1 : Read the train dataset

```python
import pandas as pd 
df = pd.read_csv("iris.csv")
df.head()
```

```python
df['species'].value_counts()
```


### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 : Seperate X and Y (specices)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 : Perform Preprocessing on X

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
```

```python

X_pre.head()
```

### Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Step 6: Create the model

```python
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier(n_neighbors=33)
model1.fit(xtrain, ytrain)
```

```python
model1.score(xtrain, ytrain)
```

```python
model1.score(xtest, ytest)
```

### Hyperparameter tuning

```python
params = {'n_neighbors':[3, 4, 5, 6, 7, 8, 9, 10, 11]}
```

```python
from sklearn.model_selection import GridSearchCV
knn = KNeighborsClassifier()
gscv1 = GridSearchCV(knn, param_grid=params, cv=5, scoring='f1_macro')
gscv1.fit(xtrain, ytrain)
```

```python
gscv1.best_params_
```

```python
gscv1.best_score_
```

```python
best_knn = gscv1.best_estimator_
best_knn
```

```python
best_knn.score(xtrain, ytrain)
```

```python
best_knn.score(xtest, ytest)
```
